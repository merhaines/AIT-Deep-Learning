```python
from PIL import Image
# Open the image
img = Image.open("ait.jpg")
# Get the size of the image
width, height = img.size
# Calculate the coordinates of the top-right 30x30 rectangle
x1 = width - 30
y1 = 0
x2 = width
y2 = 30
# Get the region of the image corresponding to the top-right 30x30 rectangle
region = img.crop((x1, y1, x2, y2))
# Fill the region with black
black = Image.new("RGB", region.size, (0, 0, 0))
img.paste(black, (x1, y1, x2, y2))
# Save the modified image
img.save("NewImage.jpg")
```
